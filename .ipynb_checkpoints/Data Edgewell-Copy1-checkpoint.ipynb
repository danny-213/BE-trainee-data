```python
import pandas as pd

# Read the inbound.csv file
df_inbound = pd.read_csv('2022 inbound.csv', sep=';')

# Remove blank "Material" rows
df_inbound = df_inbound.dropna(subset=['Material'])

#Remove blank columns
df_inbound = df_inbound.loc[:, ~df_inbound.columns.str.contains('^Unnamed')]
df_inbound = df_inbound.loc[:, ~df_inbound.columns.str.contains('^Output')]

# Find NaN values in the "pallet" column
nan_indexes = df_inbound['Pallets'].isna()

# Replace NaN values with 1
df_inbound.loc[nan_indexes, 'Pallets'] = 1


# Reset the index
df_inbound = df_inbound.reset_index(drop=True)

# Save the cleaned data to a new CSV file
df_inbound.to_csv('cleaned_inbound.csv', index=False)

# Read the MD.csv file
df_md = pd.read_csv('MD.csv', sep=';')

# Remove blank columns
df_md = df_md.dropna(axis=1, how='all')

# Save the cleaned data to a new CSV file
df_md.to_csv('cleaned_MD.csv', index=False)

df_inbound.head()
```
